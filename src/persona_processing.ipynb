```python
import numpy as np
import pandas as pd
```

```python
label = ['Freebase_char_map', 'Wikipedia Movie ID', 'Movie Name', 'char_name', 'related_name', 'Token', 'Persona_code', 'Persona_distrib']

personas_df = pd.read_csv("Data/Persona/25.100.lda.log.txt", sep='\t', names=label)

personas_df['Persona_distrib'] = personas_df['Persona_distrib'].apply(
    lambda x: np.array([float(num) for num in x.split()])
)
```

```python
personas_df.sample(5)
```

```python
label_char = ['Wikipedia Movie ID', 'Freebase Movie ID', 'Movie release date', 'Character Name', 'Actor DOB', 
                    'Actor gender', 'Actor height', 'Actor ethnicity', 'Actor Name', 'Actor age at movie release', 
                    'Freebase_char_map', 'Freebase_character_id', 'Freebase_actor_id']

char_df = pd.read_csv("Data/Persona/all.character.metadata", sep='\t', names=label_char)

char_df.head()
```

```python
char_personas = personas_df.copy()

columns_to_drop = ['Wikipedia Movie ID', 'char_name', 'related_name', 'Token', 'Persona_distrib']
char_personas = char_personas.drop(columns=columns_to_drop)

movies_char_personas = char_personas.merge(char_df, on='Freebase_char_map', how='inner')

movies_char_personas.sample(5)
```

```python
movies_char_personas.info()
```

```python
unique_movie_count = movies_char_personas['Wikipedia Movie ID'].nunique()

print(f"The data contains {len(movies_char_personas)} character personas and {unique_movie_count} different movies")
```

```python
# Save
movies_char_personas.to_csv('Data/Persona/movies_char_personas.csv', index=False)
```
