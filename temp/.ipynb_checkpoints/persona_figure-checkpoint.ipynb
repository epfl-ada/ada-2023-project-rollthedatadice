```python
import pandas as pd
import numpy as np

PATH = "../Data/Persona/out.phi.weights"

persona_verbs = dict()

with open(PATH, "r") as file:
    persona_id = 0
    verbs = pd.DataFrame(file.readline().split(sep=' '))

    for line in file.readlines():
        persona_id += 1
        freq = [float(x) for x in line.split(sep=' ')]
        top_id = np.argsort(freq)[-20:]  
        top_verbs = verbs.iloc[top_id].values.flatten().tolist()
        top_freq = [freq[i] for i in top_id]  

        persona_verbs[persona_id] = {
            'top_verbs': top_verbs,
            'top_freq': top_freq
        }

for persona_id, data in persona_verbs.items():
    print(f"Persona ID: {persona_id}")
    print(f"Top Verbs: {data['top_verbs']}")
    print(f"Top Frequencies: {data['top_freq']}")
    print()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# # 读取 CSV 文件
file_path = '../Data/Persona/persona_words.csv'
df = pd.read_csv(file_path)


persona_counts = df['persona_code'].value_counts()

persona_proportions = persona_counts / persona_counts.sum()

#df['size'] = df['persona_code'].map(persona_counts) / len(df)

persona_word = json.load(open('../Data/Persona/persona_word.json', 'r'))
categories = list(persona_word.keys())

all_words = []
for category in categories:
    all_words += persona_word[category]["top_verbs"]

all_words = list(set(all_words))

# Create a matrix where rows correspond to categories and columns to words
word_matrix = np.zeros((len(categories), len(all_words)))
for i, category in enumerate(categories):
    for word in persona_word[category]["top_verbs"]:
        word_matrix[i][all_words.index(word)] = persona_word[category]["top_freq"][persona_word[category]["top_verbs"].index(word)]

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(word_matrix)

x = pca_result[:,0]
print(x.shape)
y = pca_result[:,1]

# 创建 Dash 应用程序
app = dash.Dash(__name__)

# 页面布局
app.layout = html.Div([
    html.Div([
        dcc.Graph(id='scatter-plot')
    ], style={'width': '50%', 'display': 'inline-block'}),

    html.Div([
        dcc.Graph(id='word-freq')
    ], style={'width': '50%', 'display': 'inline-block'})
])

# 回调函数 - 更新散点图和词汇频率分布图
@app.callback(
    Output('scatter-plot', 'figure'),
    Input('scatter-plot', 'hoverData')
)
def update_scatter_plot(hover_data):
    persona_counts = df['persona_code'].value_counts()
    df['size'] = df['frequency'] / df['frequency'].sum()
    print(df['size'])

    # 创建散点图
    fig = px.scatter(df, x=x, y=y, size='size',
                     labels={'x': 'Random X', 'y': 'Random Y', 'size': 'Normalized Frequency'},
                     title='Persona Scatter Plot')
    fig.update_traces(marker={'opacity': 0.7, 'line': {'width': 0.5, 'color': 'white'}}, 
                      selector=dict(mode='markers'), hoverinfo='text')  # 悬停时显示标签
    fig.update_layout(showlegend=False)

    return fig

@app.callback(
    Output('word-freq', 'figure'),
    Input('scatter-plot', 'hoverData')
)
def update_word_freq(hover_data):
    if hover_data is None or len(hover_data['points']) == 0:
        selected_persona_id = df['persona_code'].iloc[0]
    else:
        selected_point = hover_data['points'][0]
        selected_index = selected_point['pointIndex']
        selected_persona_id = df['persona_code'].iloc[selected_index]

    # 根据选定的 persona ID 获取对应的词汇频率数据
    persona_data = df[df['persona_code'] == selected_persona_id]

    # 创建词汇频率分布图表
    fig = px.bar(persona_data, x='word', y='frequency',
                 title=f'Word Frequency for Persona ID: {selected_persona_id}',
                 labels={'word': 'Word', 'frequency': 'Frequency'})

    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import plotly.graph_objects as go

import numpy as np
np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)

f = go.FigureWidget([go.Scatter(x=x, y=y, mode='markers')])

scatter = f.data[0]
colors = ['#a3a7e4'] * 100
scatter.marker.color = colors
scatter.marker.size = [10] * 100
f.layout.hovermode = 'closest'

last_clicked = None  # To keep track of last clicked point

# create our callback function
def update_point(trace, points, selector):
    global last_clicked
    
    c = ['#a3a7e4'] * 100  # Reset colors to default for all points
    s = [10] * 100  # Reset sizes to default for all points
    
    if last_clicked is not None:
        c[last_clicked] = '#a3a7e4'  # Reset last clicked point color
        s[last_clicked] = 10  # Reset last clicked point size
    
    for i in points.point_inds:
        c[i] = '#bae2be'  # Highlight newly clicked point
        s[i] = 20
        last_clicked = i
    
    with f.batch_update():
        scatter.marker.color = c
        scatter.marker.size = s

scatter.on_click(update_point)

f
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from ipywidgets import interact


def plot_word_cloud(category):
    text = ' '.join(persona_verbs[category])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for Category {category}')
    plt.show()

interact(plot_word_cloud, category=list(persona_verbs.keys()))
```
