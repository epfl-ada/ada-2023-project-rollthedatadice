```python
import pandas as pd
import numpy as np
import json

PATH = "../Data/Persona/out.phi.weights"
```

```python
persona_verbs = dict()


with open(PATH, "r") as file:
    persona_id = 0
    verbs = pd.DataFrame(file.readline().split(sep=' '))

    for line in file.readlines(): 
        persona_id += 1
        freq = [float(x) for x in line.split(sep=' ')]
        top_id = np.argsort(freq)[-20:]
        top_verbs = verbs.iloc[top_id].values.flatten().tolist()

        #print(persona_id, top_verbs)
        persona_verbs[persona_id] = top_verbs
        
persona_verbs
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from ipywidgets import interact


def plot_word_cloud(category):
    text = ' '.join(persona_verbs[category])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for Category {category}')
    plt.show()

interact(plot_word_cloud, category=list(persona_verbs.keys()))
```
